$$
u'=4tu^{\frac{1}{2}}\\
u(0)=1\\
$$

用显示 $Ｅuler$, 隐式 $Euler$, 四阶线性多步伐, 四阶 $Runge-Kutta$ 法求解.

2.$u_{n+1}=u_n+hf(t_n,u_n)$ $\quad$ $n=0,1,2\dots,N-1$

### 龙格－库塔 runge-kutta

1.对 $t\in[0,2]$ 进行分割, 步长为 $h=0.1$
21. 对四阶龙格库塔格式为：
$$
u_{n+1} = u_n+\frac{h}{6}[k_1+2k_2+2k_3+k_4],\qquad n =0,1,2,\dots,N-1
$$
这里 
$$
k_1 = f(t_n,u_n)\\
k_2 = f(t_n+\frac{h}{2},u_n+\frac{h}{2}k_1)\\
k_3 = f(t_n+\frac{h}{2},u_n+\frac{h}{2}k_2)\\
k_4 = f(t_n+h,u_n+hk_3)
$$
